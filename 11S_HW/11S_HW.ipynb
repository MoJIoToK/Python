__15. Определим примерные корни уравнения на определенном отрезке.__

Код ниже определяет примерное значение корней на каком-то определенном заданном отрезке. В переменную `start` записываем начало интересующего отрезка, в `end` конец отрезка. Переменной `d` можно регулировать количество итераций, при `d = 1000` (малом значении) значения получаются очень грубыми, тогда как при `d = 10000` значения получаются точнее, но цена их получения увеличивается. Время работы программы увеличивается __в разы__. Насколько это целесообразно, не понятно.
Код вычисляет значение функции в точке х, и учитывается только те, которые меньше 2 (ошибка). Т.е. те значения переменной х, при подстановке которых значение функции в данной точке меньше 2. 
Вместо 2 можно поставить любое другое значение, но при этом необходимо также менять `d`. Чем меньше значение, тем больше должно быть `d`. 
С этим методом стоит быть очень осторожным и контролировать корни по графику. Т.к. при слишком большой ошибке/слишком малом значении `d` есть шанс пропустить один из корней. 

```python
start = -10
end = -3
d = 5000
for i in range(start*d, end*d):
    param = i / d
    if abs(0 - f.subs(x, param)) < 2:
        a = (i-1)/d
        b = (i + 1)/d
        print(a, b)
#         print(f.subs(x, (a + b)/2))
#        process(a, b)
```

__16. Определим корни уравнения точнее.__

Данная функция необязательна, но позволяет найти корни уравнения точнее. Для этого в переменную `a` записываем ближайшее целое меньшее значение к числу полученному в пункте 15, а в `b` ближайшее целое большее. А далее в цикле происходит перебор `i`  с уменьшением шага(дробной части), и если функция становиться больше/меньше 0, цикл прекращается. По сути код ищет переход функции оси Х.
Единственное что нужно менять знак в условии. `<` - если график пересекает ось Х сверху (уменьшается), `>` - если график пересекает ось Х снизу (увеличивается)

```python
a = -5
b = -4
d = 1
for _ in range(20):
    a *= 10
    b *= 10
    d *= 10
    
    for i in range(a, b):
        param = i / d
        if y.subs(x, param) < 0: # < - если график пересекает ось Х сверху (уменьшается), > - если график пересекает ось Х снизу (увеличивается)
            a = i - 1
            b = i + 1
            break
print(a / d, b / d)

print(y.subs(x, a / d))
print(y.subs(x, b / d))
print(y.subs(x, (a / d + b / d)/2))

        #print(param, f.subs(x, param))
```
